- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q1
list_over25=[]
list_over10=[]
list_else_str=[]
list_else=[]

for i in range(54):
    if df[df.columns[i]].isnull().sum()/len(df[df.columns[i]])>=0.25:
        list_over25.append(df.columns[i])
    elif df[df.columns[i]].isnull().sum()/len(df[df.columns[i]])>=0.1:
        list_over10.append(df.columns[i])
    elif df[df.columns[i]].dtype=='object':
        list_else_str.append(df.columns[i])
    else:
        list_else.append(df.columns[i])
    continue

df.drop(list_over25,axis=1,inplace=True)
df.dropna(subset=list_over10,inplace=True)
for i in list_else_str:
    df[i]=df[i].fillna(df[i].value_counts().idxmax())
    continue
for i in list_else:
    df[i]=df[i].fillna(df[i].median())
    continue
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1) 수용인원 변수는 이용목적에 따라 구분될 수 있도록 범주형으로 전환
accommodates_group=[]
for a in df['accommodates']:
    if (0<=a<5):
        accommodates_group.append('couple/family')
    elif (5<=a<10):
        accommodates_group.append('small group')
    elif (20<=a<20):
        accommodates_group.append('big group')
    else:
        accommodates_group.append('20 or more')
df['accommodates_group']=accommodates_group
df.head()
```

```python
#Q13-2) 가격책정에 중요 요소인 범주형 변수를 수치화해주기 위해 원핫인코딩 진행
room_type_encoded=pd.get_dummies(df['room_type'])
room_type_encoded
```

```python
#Q13-3) 수치형 변수들 리스케일링을 통해 관계 도출의 용이성 제고
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()

mm_scaler1=mm_scaler.fit_transform(df['price'].values.reshape(-1,1))
df['mm_scaled_price']=mm_scaler1

mm_scaler2=mm_scaler.fit_transform(df['number_of_reviews'].values.reshape(-1,1))
df['mm_scaled_num_of_reviews']=mm_scaler2

df.head()
```
