#### birthDate 컬럼으로 age 컬럼 추가하기

1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
players['birthDate']=pd.to_datetime(players['birthDate'])
players.info()
```

```python
players.head(3)
```

2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산




```python
#한국나이계산

from datetime import datetime

players['age']=players['birthDate'].apply(lambda x: (datetime.now().year)+1-(x.year))

players.head(30)
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
players['age']=players['age'].fillna(players['age'].mean())
players.tail(30)
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
players['heightCm']=players['height'].apply(lambda x: float(x.split('-')[0])*30.48 +float(x.split('-')[1])*2.54)
players.head(30)
```
