#### 정규방정식

```python
import numpy as np
from numpy.linalg import inv
```

```python
def estimate_beta(X,y):
    designX=np.c_[np.ones((9,1)),X]
    beta_hat=np.linalg.inv(designX.T.dot(designX)).dot(designX.T).dot(y)
    return beta_hat
```

```python
beta_hat = estimate_beta(X, y)
beta_hat
```

```python
X_new = np.array([[40,60,4,900,60]])
X_new
```

```python
X_new_b = np.c_[np.ones((1,1)), X_new]
predict = X_new_b.dot(beta_hat)
predict
```

#### 실제 값과 비교

```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)

pred = model.predict(X_new)
```

```python
print(model.intercept_)   # intercept
print(model.coef_)   # 추정된 회귀계수 (intercept 제외)
print(pred)
```
